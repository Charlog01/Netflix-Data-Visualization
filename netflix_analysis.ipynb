# Netflix Data Visualization Project

```r
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```r
Data Preparation
```

```r
old_name = "Netflix_data.csv"
new_name = "Netflix_shows_movies.csv"
os.rename(old_name, new_name)
```

```r
df=pd.read_csv('Netflix_shows_movies.csv')
df.head()
```

```r
Data Cleaning
```

```r
df.isnull().sum()
```

```r
Data Exploration
```

```r
genres=df['listed_in'].str.split(', ',expand=True).stack().value_counts().head(10)
genres
```

Data Visualization

```r
Most watched genres
```

```r
plt.figure(figsize=(10,5))
sns.barplot(x=genres.values,y=genres.index)
plt.show()
```

```r
Ratings distribution
```

```r
plt.figure(figsize=(10,5))
sns.countplot(y=df['rating'],order=df['rating'].value_counts().index)
plt.show()
```
